```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#binomial distro
n, p = 10, .5  # number, prob
bi_hist = np.random.binomial(n, p, 1000)
```

```python
count, bins, ignored = plt.hist(bi_hist, 14, density=True)
plt.show()
```

```python
#normal distro
mn, stddev = 0, 0.1 # mean and standard deviation
n_hist = np.random.normal(mn, stddev, 1000)
```

```python
count, bins, ignored = plt.hist(n_hist, 14, density=True)
plt.show()
```

```python
#poisson distro
p_hist = np.random.poisson(5, 10000)
```

```python
count, bins, ignored = plt.hist(p_hist, 14, density=True)
plt.show()
```
